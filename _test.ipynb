```python
#RSA
def get_e(n,phi):
   # n=p*q
    #phi=(p-1)*(q-1)
    for e in range(2,phi):
        coprime=True
        for x in range(2,e+1):
            if (e%x==0 and n%x==0 ) or (e%x==0 and phi%x==0):
                coprime=False
                break
        if coprime:
            return e
    
def get_d(e,phi):
    d=1
    while (d*e)%phi!=1 or d==e:
        d+1
    return d
    
    
def get_rsa_keys(p,q):
    n=p*q
    phi=(p-1)*(q-1)
    e=get_e(n,phi)
    d=get_d(e,phi)
    public_key=(e,n)
    private_key=(d,n)
    return public_key,private_key

def encrypt (srcpath, dstpath , public_key):
    e,n =public_key
    with open(srcpath,'rb') as f:
        plaintext=f.read()
    ciphertext=bytes([(p**e)%n for p in plaintext])
    with open (dstpath , 'wb') as f:
        f.write(ciphertext)
    
def decrypt (srcpath, dstpath , private_key):
    d,n =private_key
    with open(srcpath,'rb') as f:
        ciphertext=f.read()
    plaintext=bytes([(c**d)%n for c in ciphertext])
    with open (dstpath , 'wb') as f:
        f.write(plaintext)
        
        
```

```python
#play fair 
import numpy as np
def get_matrix(key):
    key+='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    key=key.replace('J','I')
    matrix=[]
    for k in key:
        if k not in matrix:
            matrix.append(k)
    return np.array(matrix).reshape(5,5)
def get_indices(matrix):
    indices={}
    for i in range(5):
        for j in range(5):
            letter=matrix[i,j]
            indices[letter]=(i,j)
    return indices
def get_pairs(text):
    text=text.replace('J','I')
    result=''
    #########
    for i in range(len(text)-1):
        result+=text[i]
        if text[i]==text[i+1] and len(result)%2==1:
            result+='X'
    result+=text[-1]
    ########
    if len(result)%2==1:
        result+='X'
    #######
    pairs=[]
    for i in range(0,len(result)-1,2):
        pairs.append(result[i]+result[i+1])  ### +
    return pairs

def encrypt(plaintext,key):
    matrix=get_matrix(key)
    indices=get_indices(matrix)
    pairs=get_pairs(plaintext)
    ciphertext=''
    for a,b in pairs:
        r1,c1=indices[a]
        r2,c2=indices[b]
        #####
        if r1==r2:
            c1=(c1+1)%5
            c2=(c2+1)%5
        elif c1==c2:
            r1=(r1+1)%5
            r2=(r2+1)%5
        else:
            c1,c2=c2,c1
        a=matrix[r1,c1]
        b=matrix[r2,c2]
        ciphertext+=a+b
    return ciphertext

def decrypt(ciphertext,key):
    matrix=get_matrix(key)
    indices=get_indices(matrix)
    pairs=get_pairs(ciphertext)
    plaintext=''
    for a,b in pairs:
        r1,c1=indices[a]
        r2,c2=indices[b]
        #####
        if r1==r2:
            c1=(c1-1)%5
            c2=(c2-1)%5
        elif c1==c2:
            r1=(r1-1)%5
            r2=(r2-1)%5
        else:
            c1,c2=c2,c1
        a=matrix[r1,c1]
        b=matrix[r2,c2]
        plaintext+=a+b
    return plaintext
    
    
#matrix=get_matrix("PLYFAIR")
#ind=get_indices(matrix)
#print(ind)
#p=get_pairs("salmaaa")
#print(p)
en=encrypt("SALMA","PLAYFAIR")
print(en)
de=decrypt(en,"PLAYFAIR")
print(de)
```
